```python
rcmd = rcmd_60min.copy()
rcmd['grade'] = rcmd['relative_score'] // 30 +1
```

```python
bins = np.arange(0,1100,100)
labels = [str(i) for i in bins[1:]]
rcmd['rcmd_cnt_bin'] = pd.cut(rcmd['rcmd_cnt'],bins=bins,labels=labels,include_lowest=True)

df_max_rcmd = rcmd.sort_values(by=['user_id', 'rcmd_cnt'], ascending=[True, False])
rcmd_unique = df_max_rcmd.drop_duplicates(subset=['user_id'], keep='first')
pivot_rcmd = rcmd.pivot_table(index = 'grade',columns='rcmd_cnt_bin',values='user_id',aggfunc='nunique',fill_value=0,observed=False)
pivot_rcmd
```

```python
pivot_rcmd = rcmd_unique.pivot_table(index = 'grade',columns='rcmd_cnt_bin',values='user_id',aggfunc='nunique',fill_value=0,observed=False)
pivot_rcmd
```
